```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('C:/Users/Dell/Desktop/churn-customer-diagnosis/data/processed/telco_churn_cleaned.csv')
print("Shape:", df.shape)
```

### Contract type vs Churn (Contract type and Churn are independent or related)


```python
contingency = pd.crosstab(df['Contract'], df['Churn'])
print(contingency)
chi2, p_value, dof, expected = stats.chi2_contingency(contingency)
print(f"\nChi-square statistic: {chi2:.2f}")
print(f"p-value: {p_value:.10f}")
print(f"Degrees of freedom: {dof}")
if p_value < 0.05:
    print("\n Result: REJECT null hypothesis — Contract type IS significantly related to Churn")
else:
    print("\n Result: FAIL to reject null — no significant relationship found")
```

### Tenure vs Churn (tenure is the SAME for churned and retained customers or Different)

```python
tenure_churned = df[df['Churn'] == 'Yes']['tenure']
tenure_retained = df[df['Churn'] == 'No']['tenure']

t_stat, p_value = stats.ttest_ind(tenure_churned, tenure_retained)

print(f"Mean tenure (Churned): {tenure_churned.mean():.1f} months")
print(f"Mean tenure (Retained): {tenure_retained.mean():.1f} months")
print(f"\nt-statistic: {t_stat:.2f}")
print(f"p-value: {p_value:.10f}")
if p_value < 0.05:
    print("\n Result: REJECT null — tenure IS significantly different between churned and retained customers")
else:
    print("\n Result: FAIL to reject null — no significant difference found")
```

### Monthly Charges vs Churn 

```python
charges_churned = df[df['Churn'] == 'Yes']['MonthlyCharges']
charges_retained = df[df['Churn'] == 'No']['MonthlyCharges']

t_stat, p_value = stats.ttest_ind(charges_churned, charges_retained)

print(f"Mean Monthly Charges (Churned): ${charges_churned.mean():.2f}")
print(f"Mean Monthly Charges (Retained): ${charges_retained.mean():.2f}")
print(f"\nt-statistic: {t_stat:.2f}")
print(f"p-value: {p_value:.10f}")

if p_value < 0.05:
    print("\n Result: REJECT null — Monthly Charges IS significantly different between groups")
else:
    print("\n Result: FAIL to reject null")
```

### Internet Service type vs Churn 

```python
contingency2 = pd.crosstab(df['InternetService'], df['Churn'])
print(contingency2)
chi2, p_value, dof, expected = stats.chi2_contingency(contingency2)

print(f"\nChi-square statistic: {chi2:.2f}")
print(f"p-value: {p_value:.10f}")

if p_value < 0.05:
    print("\n Result: REJECT null — Internet Service type IS significantly related to Churn")
else:
    print("\n Result: FAIL to reject null")
```

```python
summary = pd.DataFrame({
    'Hypothesis': [
        'Contract Type vs Churn',
        'Tenure vs Churn',
        'Monthly Charges vs Churn',
        'Internet Service vs Churn'
    ],
    'Test Used': ['Chi-square', 't-test', 't-test', 'Chi-square'],
    'Significant (p<0.05)?': ['Yes', 'Yes', 'Yes', 'Yes']
})
print(summary.to_string(index=False))
```
